```python
# Your answer here

from typing import List
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val = 0, left = None, right = None):
        self.val = val
        self.left = left
        self.right = right

def bt_path(root: TreeNode) -> List[List[int]]:
    
    if not root:
        return root
    
    ans = []
    curr_path = []
    bt_path_helper(root, curr_path, ans)

    return ans

def bt_path_helper(node, curr_path, ans):
    if not node.left and not node.right:
        curr_path = curr_path + [node.val]
        ans.append(curr_path)
        return 
    
    if node.left:
        curr_path = curr_path + [node.val]
        bt_path_helper(node.left, curr_path, ans)
        curr_path.pop()
        
    if node.right:
        curr_path = curr_path + [node.val]
        bt_path_helper(node.right, curr_path, ans)
        curr_path.pop()
    

# Test case 1
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(2)
root.left.left = TreeNode(3)
root.left.right = TreeNode(5)
root.right.left = TreeNode(6)
root.right.right = TreeNode(7)

print(bt_path(root)) # Output: [[1, 2, 3], [1, 2, 5], [1, 2, 6], [1, 2, 7]]



# Test case 2
root = TreeNode(10)
root.left = TreeNode(9)
root.right = TreeNode(7)
root.left.left = TreeNode(8)

print(bt_path(root)) # Output: [[10, 9, 8], [10, 7]] 
```

The time complexity of this solution is O(N) because it visits each node exactly once using depth-first search (DFS) to explore all possible root-to-leaf paths. Since every node is processed once, the traversal remains efficient. 

The space complexity depends on two factors: recursion depth and path storage. The recursion depth is at most the height of the tree, which is O(H), where H is log(N) for a balanced tree but can be O(N) for a skewed tree. Additionally, storing all root-to-leaf paths in the result list requires space. Thus, while the function runs in linear time, its space usage can grow significantly when there are many paths to store.
